```python
import random
import matplotlib.pyplot as plt

# Define the probabilities and distances for each direction
probabilities = {'F': 0.40, 'B': 0.10, 'L': 0.25, 'R': 0.25}
distances = {'F': 75, 'B': 45, 'L': 60, 'R': 60}

def move(direction, x, y):
    distance = distances[direction]
    if direction == 'F':
        y += distance
    elif direction == 'B':
        y -= distance
    elif direction == 'L':
        x -= distance
    elif direction == 'R':
        x += distance
    return x, y

def simulate_walk(random_numbers):
    x, y = 0, 0
    x_positions = [x]
    y_positions = [y]
    print("Step Random Numbers Direction X Coordinate Y Coordinate")
    for i, num in enumerate(random_numbers, start=1):
        direction = 'F' if num < 4 else 'B' if num == 4 else 'L' if num <= 6 else 'R'
        x, y = move(direction, x, y)
        x_positions.append(x)
        y_positions.append(y)
        print(f"{i}             {num}          {direction}           {x}           {y}")
    return x_positions, y_positions

def main():
    random_numbers = [6, 2, 0, 6, 8, 5, 7, 7, 9, 8, 4, 8, 2, 6, 2, 1, 3, 9, 8, 4]
    x_positions, y_positions = simulate_walk(random_numbers)

    plt.plot(x_positions, y_positions, marker='o')
    plt.title('Random Walk')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
